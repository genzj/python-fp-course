## 设计一个模块，基于斐波那契数列，实现以下所有功能：

- 数列前N项
- 数列第N至M项
- 值为奇数的N项
- 下标为奇数的N项


### 无模式——大量重复代码

```python
def fib_first_n(n):
  answer = [1, 1]

  while len(answer) < n:
    answer.append(sum(answer[-2:]))
  
  return answer[:n]


fib_first_n(10)
```

```python
def fib_n_to_m(n, m):
  answer = [1, 1]

  while len(answer) < m:
    answer.append(sum(answer[-2:]))
  
  return answer[n:m]


fib_n_to_m(5, 10)
```

```python
def fib_odd_value_n(n):
  answer = [1, 1]
  last_two = list(answer)

  while len(answer) < n:
    last_two = [last_two[-1], sum(last_two)]
    if last_two[-1] % 2 != 0:
      answer.append(last_two[-1])
  
  return answer[:n]


fib_odd_value_n(7)
```

```python
def fib_odd_index_n(n):
  answer = [1, 1]

  while len(answer) < 2*n:
    answer.append(sum(answer[-2:]))
  
  return answer[1:2*n:2]


fib_odd_index_n(5)
```

### 面向对象模式

- 较多结构代码
- 自由化的状态保存

```python
class Fib:
  def __init__(self):
    self.last_two = []
  
  def next(self):
    if len(self.last_two) < 2:
      p = 1
    else:
      p = sum(self.last_two)
    self.last_two = self.last_two[-1:] + [p,]

    return p


fib = Fib()
[fib.next() for _ in range(10)]
```

```python
def fib_first_n_oo(n):
  fib = Fib()
  answer = []
  for idx in range(n):
    v = fib.next()
    answer.append(v)
  return answer


fib_first_n_oo(10)
```

```python
def fib_n_to_m_oo(n, m):
  fib = Fib()
  answer = []
  for idx in range(m):
    v = fib.next()
    if idx >= n:
      answer.append(v)
  return answer


fib_n_to_m_oo(5, 10)
```

```python
def fib_odd_value_n_oo(n):
  fib = Fib()
  answer = []
  while len(answer) < n:
    v = fib.next()
    if v % 2 != 0:
      answer.append(v)
  return answer


fib_odd_value_n_oo(7)
```

```python
def fib_odd_index_n_oo(n):
  fib = Fib()
  answer = []
  for idx in range(2*n):
    v = fib.next()
    if idx % 2 != 0:
      answer.append(v)
  return answer

fib_odd_index_n_oo(5)
```

### 生成器模式

- 减少实现和实例化类的成本（状态保存自动化）
- 代码风格更统一

```python
def generate_fib():
  yield 1
  yield 1
  last_two = [1, 1]
  while True:
    v = sum(last_two)
    yield v
    last_two = [last_two[-1], v]
```

```python
def fib_first_n_gen(n):
  for idx, v in enumerate(generate_fib()):
    if idx >= n:
      break
    yield v


list(fib_first_n_gen(10))
```

```python
def fib_n_to_m_gen(n, m):
  for idx, v in enumerate(generate_fib()):
    if idx < n:
      continue
    if idx >= m:
      break
    yield v


list(fib_n_to_m_gen(5, 10))
```

```python
def fib_odd_value_n_gen(n):
  cnt = 0
  for v in generate_fib():
    if cnt >= n:
      break
    if v % 2 == 0:
      continue
    yield v
    cnt += 1


list(fib_odd_value_n_gen(7))
```

```python
def fib_odd_index_n_gen(n):
  cnt = 0
  for idx, v in enumerate(generate_fib()):
    if cnt >= n:
      break
    if idx % 2 == 0:
      continue
    yield v
    cnt += 1


list(fib_odd_index_n_gen(5))
```
