## 流式数据处理

- 循环被map操作替代。
- 结果汇聚等终止操作由reduce操作完成。
- 中间可以加入更多操作，如过滤（filter）、副作用等。


### 例 —— 统计有5个以上英文字符的词

```python
words = '''
The European languages are members of the same family.
Their separate existence is a myth.
For science, music, sport, etc, Europe uses the same vocabulary.
'''.split()

print(words)
```

```python
def stem(word):
  """ 去除标点符号 """
  return word.lower().rstrip(",.!:;'-\"").lstrip("'\"")
```

```python
import functools

functools.reduce(
  lambda cnt, word: cnt + 1,
  filter(
    lambda word: len(word) > 5,
    map(
      stem, words
    )
  ),
  0
)
```

```python
import toolz

toolz.reduce(
  lambda cnt, word: cnt + 1,
  toolz.pipe(
    words,
    toolz.curried.map(stem),
    toolz.curried.filter(lambda word: len(word) > 5)
  ),
  0
)
```

```python
filter_stem_gt_5 = toolz.compose(
  toolz.curried.filter(lambda word: len(word) > 5),
  toolz.curried.map(stem),
)

toolz.reduce(
  lambda cnt, word: cnt + 1,
  filter_stem_gt_5(words),
  0
)
```
