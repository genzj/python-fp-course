## For循环（迭代）的底层逻辑

```python
for i in range(5):
  print(i)
```

For循环（迭代）基本等价于下面的While循环

```python
it = iter(range(5))
while True:
  try:
    i = next(it)
  except StopIteration:
    break
  print(i)
```

而`next`等价于调用`__next__`成员函数；`iter`等价于调用`__iter__`成员函数。

```python
it = range(5).__iter__()
while True:
  try:
    i = it.__next__()
  except StopIteration:
    break
  print(i)
```

```python
class MyRange:
  def __init__(self, n):
    self.i = 0
    self.n = n
  
  def __next__(self):
    i = self.i
    if i >= self.n:
      raise StopIteration()
    self.i += 1
    return i

  def __iter__(self):
    return self


for i in MyRange(5):
  print(i)
```

## 实验

实现一个“对折”迭代器，给定初值，每次迭代输出上一次迭代值的一半，直至值小于1e-9

```python
class Half:
  def __init__(self, initial) -> None:
    self.n = float(initial)
  
  def __next__(self):
    raise StopIteration()
  
  def __iter__(self):
    return self
```

```python
for i in Half(1024):
  print(i)
```

### 参考实现

```python
class Half:
  def __init__(self, initial) -> None:
    self.n = float(initial)
  
  def __next__(self):
    n = self.n
    if n < 1e-9:
        raise StopIteration()
    self.n /= 2
    return n
  
  def __iter__(self):
    return self


for i in Half(1024):
  print(i)
```
